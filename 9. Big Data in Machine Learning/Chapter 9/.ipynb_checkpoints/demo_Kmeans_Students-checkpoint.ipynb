```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('kmeans_demo').getOrCreate()
```

```python
# Loads data.
data = spark.read.csv("5000_points.txt", header=False,
                      inferSchema=True, sep="\t")
```
