## Chuẩn bị, chuẩn hóa dữ liệu, xác định input, output

```python
csv_file_path = "Du lieu cung cap/tmdb_5000_movies.csv"
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import from_json
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, ArrayType
from pyspark.sql.functions import expr

# Read the CSV file into a DataFrame
data = spark.read.csv(csv_file_path, header=True, inferSchema=True, quote='"', escape='"', multiLine=True)

genres_schema = ArrayType(StructType([
    StructField("id", IntegerType()),
    StructField("name", StringType())
]))

keywords_schema = ArrayType(StructType([
    StructField("id", IntegerType()),
    StructField("name", StringType())
]))

production_companies_schema = ArrayType(StructType([
    StructField("id", IntegerType()),
    StructField("name", StringType())
]))

production_countries_schema = ArrayType(StructType([
    StructField("iso_3166_1", IntegerType()),
    StructField("name", StringType())
]))


spoken_languages_schema = ArrayType(StructType([
    StructField("iso_639_1", IntegerType()),
    StructField("name", StringType())
]))

# Extract values from the nested JSON array in the genres column
data = data.withColumn("genres", from_json(data["genres"], genres_schema))
data = data.withColumn("keywords", from_json(data["keywords"], keywords_schema))
data = data.withColumn("production_companies", from_json(data["production_companies"], production_companies_schema))
data = data.withColumn("production_countries", from_json(data["production_countries"], production_countries_schema))
data = data.withColumn("spoken_languages", from_json(data["spoken_languages"], spoken_languages_schema))
```

### Feature Transformations

```python
from pyspark.ml.feature import Tokenizer,StopWordsRemover
from pyspark.ml.feature import CountVectorizer, IDF, StringIndexer
tokenizer = Tokenizer(inputCol="text", outputCol="token_text")
stopremove = StopWordsRemover(inputCol='token_text',outputCol='stop_tokens')
count_vec = CountVectorizer(inputCol='stop_tokens',outputCol='c_vec')
idf = IDF(inputCol="c_vec", outputCol="tf_idf")
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.linalg import Vector
```

```python
clean_up = VectorAssembler(inputCols=['tf_idf','length'], outputCol='features')
```

### Pipeline

```python
from pyspark.ml import Pipeline
```

```python
data_prep_pipe = Pipeline(stages=[tokenizer,
                                  stopremove,
                                  count_vec,
                                  idf,
                                  clean_up])
```

```python
cleaner = data_prep_pipe.fit(data)
```

```python
clean_data = cleaner.transform(data)
```

### The Model Kmean

```python
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator
```

```python
# Trains a k-means model.
k_list = []
silhouette_list = []
for k in range(2,11): 
    kmeans = KMeans(featuresCol='features',k=k)
    model = kmeans.fit(clean_data)
    predictions  = model.transform(clean_data)
    evaluator = ClusteringEvaluator()
    silhouette = evaluator.evaluate(predictions)
    k_list.append(k)
    silhouette_list.append(silhouette)
    print("With k =", k, "Set Sum of Silhouette = " + str(silhouette))
```

```python
plt.plot(k_list, silhouette_list)
plt.show()
```

- Chọn k = 5

```python
num_clusters = 5
```

```python
# Trains a k-means model.
kmeans = KMeans(featuresCol='features',k=num_clusters)
model = kmeans.fit(clean_data)
```

```python
# silhoutte
predictions = model.transform(clean_data)
# Evaluate clustering by computing Silhouette score
evaluator = ClusteringEvaluator()
silhouette = evaluator.evaluate(predictions)
print("Silhoutte = " + str(silhouette))
```

```python
# Shows the result.
centers = model.clusterCenters()
print("Cluster Centers: ")
for center in centers:
    print(center)
```

```python
predictions = model.transform(clean_data)
```

```python
# Display main characteristics of each cluster
for cluster_id in range(num_clusters):
    cluster_movies = predictions.filter(f'prediction = {cluster_id}')
    #print(f"\nCluster {cluster_id + 1} - Movie Count: {cluster_movies.count()}")
    #cluster_movies.select('text').show(truncate=False)

    # Generate Word Cloud for the cluster
    cluster_text = ' '.join([row['text'] for row in cluster_movies.collect()])
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(cluster_text)

    # Plot the Word Cloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Cluster {cluster_id + 1} - Word Cloud')
    plt.show()
```
