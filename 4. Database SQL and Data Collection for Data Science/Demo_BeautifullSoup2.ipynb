```python
from bs4 import BeautifulSoup
import requests
```

```python
page_url = "https://dienmaythiennamhoa.vn/tivi.html"
```

```python
page_reponse = requests.get(page_url)
```

```python
page_content = BeautifulSoup(page_reponse.content,"html.parser")
```

```python
page_reponse.content
```

```python
product_name = page_content.find_all(name = 'div', attrs = {'class':'title-card'})
```

```python
len(product_name)
```

```python
lst_names = []
for product in product_name:
    lst_names.append(product.text)
```

```python
lst_names[:5]
```

```python
## ly ds gi

product_prices = page_content.find_all(name = 'p', attrs = {'class':'sale-price product-sale-price'})
```

```python
len(product_prices)
```

```python
lst_price = []
for product in product_prices:
    lst_price.append(product["data-price"])
```

```python
lst_price[:5]
```

```python
lst_price = []
for product in product_prices:
    lst_price.append(float(product.text.strip().replace('','')))
```

```python
lst_price[:5]
```

```python
# ly ds hnh sn phm
product_images = page_content.find_all(name = 'img', attrs = {'class':'lazyload img-responsive'})

```

```python
len(product_images)
```

```python
lst_images = []
for product in product_images:
    lst_images.append(product["data-src"].split('/')[-1])
```

```python
lst_images[:5]
```

```python
# Ti file hnh v lu vo th mc image
```

```python
for product in product_images:
    link = product["data-src"]
    file_name = link.split('/')[-1]
    with open('images/' +  file_name, 'wb') as f:
        image_response = requests.get(link)
        f.write(image_response.content)
```

```python
import pandas as pd
```

```python
# to df
df = pd.DataFrame({
    'name': lst_names,
    'price': lst_price,
    'image':  lst_images
})
df.head()
```

```python
df.to_csv('tivi.csv', index=False)
```
