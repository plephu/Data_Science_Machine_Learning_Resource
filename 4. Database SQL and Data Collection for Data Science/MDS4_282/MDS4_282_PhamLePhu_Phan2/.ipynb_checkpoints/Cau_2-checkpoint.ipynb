```python
%%sql result <<
select concat(FirstName,' ', LastName) as customer_name, count(*) as count_orders
from customers c , orders o 
where c.Id = o.CustomerId
group by customer_name
order by count_orders desc
limit 10
```

```python
result
```

```python
df = pd.DataFrame(result)
df.head(10)
```

```python
def chart1(df:pd.DataFrame, x_col:str, y_col:str, title:str, x_title:str, y_title:str) -> None:
    plt.figure(figsize=(8,6))
    sns.barplot(data=df, x=x_col, y=y_col)
    plt.title(title, fontsize=10, color='red')
    plt.xlabel(x_title)
    plt.ylabel(y_title)
    plt.show()
```

```python
%%fsql
SELECT *
FROM df
OUTPUT USING chart1(x_col='count_orders', 
                    y_col='customer_name', 
                    title='10 kh치ch h맕g c칩 nhi敲u 캠쿼n 캟故읎 h맕g nh故볎',
                    x_title='S敲 캠쿼n 캠故읎 h맕g', 
                    y_title='T칡n kh치ch h맕g')
```

```python
%%sql result1 <<
SELECT YEAR(o.OrderDate) as YEAR, SUM(o.TotalAmount) as sum_TotalAmount
FROM orders o
GROUP BY YEAR(o.OrderDate)
ORDER BY YEAR(o.OrderDate);
 
```

```python
%%sql result2 <<
SELECT YEAR(o.OrderDate) as YEAR, count(o.id) count_order
FROM orders o
GROUP BY YEAR(o.OrderDate)
ORDER BY YEAR(o.OrderDate);
```

```python
df1 = pd.DataFrame(result1)
df1 = df1.astype(int)
df1.head()
```

```python
df2 = pd.DataFrame(result2)
df2.head()
```

```python
plt.figure(figsize=(15,10))
fig, axs = plt.subplots(ncols=2)
plt.title("Th敲녍g k칡 theo n캒m", fontsize=10, color='blue')

sns.barplot(data=df1, x='YEAR', y='sum_TotalAmount',ax=axs[0])
axs[0].set_title('Th敲녍g k칡 theo tr敲 gi치 캠쿼n 캠故읎 h맕g')

axs[0].set_xlabel('N캒m',color='red') 
axs[0].set_ylabel('T敲뎙g tr敲 gi치',color='red') 

sns.barplot(data=df2, x='YEAR', y='count_order',ax=axs[1])
axs[1].set_title('Th敲녍g k칡 theo s敲 캠쿼n 캠故읎 h맕g') 
axs[1].set_xlabel('N캒m',color='red')
axs[1].set_ylabel('S敲 캠쿼n 캠故읎 h맕g',color='red') 


plt.show()
```
