```python
%%sql SELECT country, AVG(unitprice) AS avg_unitprice, COUNT(p.Id) AS count_product, COUNT(distinct sp.City) AS count_city
FROM products p, suppliers sp
WHERE p.SupplierId = sp.id AND p.IsDiscontinued = 0
GROUP BY country
ORDER BY count_product ASC, avg_unitprice DESC;
```

```python
%%sql
SELECT c.FirstName, c.LastName, o.OrderNumber, o.OrderDate, o.TotalAmount
FROM ORDERs o, customers c
WHERE o.CustomerId = c.Id
AND MONTH(o.OrderDate) = 3 AND YEAR(o.OrderDate) = 2014
ORDER BY o.OrderDate DESC, o.TotalAmount desc;
```

```python
%%sql
SELECT a.order_year, a.ProductName, a.total_amount 
FROM ( 
SELECT YEAR(o.OrderDate) AS order_year, ProductName, 
SUM(ot.UnitPrice * ot.Quantity) AS total_amount, 
rank() over (PARTITION BY YEAR(o.OrderDate) ORDER BY SUM(ot.UnitPrice * ot.Quantity)) AS rank_order
FROM (orderitems ot JOIN products p ON ot.ProductId = p.Id)
JOIN orders o ON o.Id = ot.OrderId
GROUP BY order_year, ProductName
ORDER BY total_amount) a
WHERE rank_order=1
ORDER BY a.order_year;
```

```python
%%sql
select concat(FirstName,' ', LastName) as customer_name, Country, City, count(*) as count_orders
from customers c , orders o 
where c.Id = o.CustomerId
group by customer_name, Country, City
having count_orders > 15
order by count_orders desc
```

```python
%%sql
SELECT p.ProductName, ifnull(SUM(i.Quantity),0) AS sum_quantity
FROM products p LEFT JOIN orderitems i ON p.id=i.ProductId LEFT JOIN orders o ON o.id=i.OrderId
WHERE p.IsDiscontinued = 0
GROUP BY p.ProductName
ORDER BY sum_quantity 
```
