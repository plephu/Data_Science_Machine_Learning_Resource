### Tạo DataFrame từ kết quả trả về 

#### Cách 1: dùng pd.DataFrame

```python
import pandas as pd
```

```python
# hiển thị các khách hàng ở Mexico và lưu vào biến
results = %sql Select * From Customers where country = 'mexico'
```

```python
df = pd.DataFrame(results)
df
```

#### Cách 2: dùng pd.read_sql(câu lệnh sql, biến kết nối)

```python
# tạo kết nối
from sqlalchemy.engine import create_engine
```

```python
cn = create_engine("mysql+pymysql://root:@localhost/productshipping?charset=utf8")
```

```python
sql = "Select * From Customers where country = 'mexico'"
df = pd.read_sql(sql, cn)
```

```python
df
```

#### sử dụng tham số trong magic sql

```python
# hiển thị các khách hàng ở quốc gia do người dùng truyền vào
pCountry = input('Nhập tên quốc gia: ')
```

```python
%sql Select * From Customers where country = :pCountry
```

### Thực thi câu lệnh truy vấn trên nhiều dòng (sử dụng %%sql)

```python
%%sql 
Select * 
From Customers 
Where country = 'mexico'
```

```python
#lưu kết quả vào biến trong trường hợp sử dụng %sql
results = %sql Select * From Customers Where country = 'mexico'
```

```python
results
```
