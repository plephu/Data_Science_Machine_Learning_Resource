### Handle Outlier

#### Keep all outlier

```python
df_titanic_in_keep = pd.concat([df_titanic_in_category, df_titanic_in_numeric, df_titanic_out], axis=1)
```

```python
df_titanic_in_keep.head()
```

```python
df_titanic_in_keep.isnull().sum()
```

#### Remove outlier

```python
df_titanic_in_remove = pd.concat([df_titanic_in_category, df_titanic_in_numeric, df_titanic_out], axis=1)
```

```python
_analyzer.check_outlier_of_numerical_variable('Age', df_titanic_in_remove)
```

```python
union_outlier = []
for col in df_titanic_in_numeric.columns:
    outliers = _analyzer.check_outlier_of_numerical_variable(col, df_titanic_in_remove)
    if outliers:
        upper, lower = outliers
        union_outlier += upper.to_list()
        union_outlier += lower.to_list()
outlier_index = pd.Index(list(set(union_outlier)))
```

```python
df_titanic_in_remove.drop(outlier_index)
```

```python
upper_outlier = []
lower_outlier = []
for col in df_titanic_in_numeric.columns:
    outliers = _analyzer.check_outlier_of_numerical_variable(col, df_titanic_in_remove)
    if outliers:
        upper, lower = outliers
        upper_outlier += upper.to_list()
        lower_outlier += lower.to_list()
outlier_index = pd.Index(list(set(upper_outlier).intersection(set(lower_outlier))))
```

```python
df_titanic_in_remove.drop(outlier_index, inplace=True)
```
