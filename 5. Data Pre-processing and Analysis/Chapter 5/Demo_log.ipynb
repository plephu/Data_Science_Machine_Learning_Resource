```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./input_data/wine_types.csv')
df.head()
```

```python
### before apply log
print(f'Varian of Alcohol: {df.Alcohol.var()}')
print(f'Varian of Proline: {df.Proline.var()}')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.distplot(df.Proline)
plt.show()
```

```python
df['Proline_log'] = np.log(df.Proline)
```

```python
### After apply log
print(f'Varian of Proline after: {df.Proline_log.var()}')
```

```python
sns.distplot(df.Proline_log)
plt.show()
```

```python
f, (ax1, ax2) = plt.subplots(1, 2)
sns.distplot(df.Proline_log, ax=ax1)
df.Proline_log.plot.hist(ax=ax2)
plt.tight_layout()
plt.show()
```
