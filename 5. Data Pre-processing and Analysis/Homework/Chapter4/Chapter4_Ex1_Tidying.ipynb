```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

%config Completer.use_jedi = False
```

```python
df= pd.read_excel('input_data/BMI.xlsx')
df
```

```python
# Melt df into new dataframe: df_melted
df_melted=pd.melt (df, id_vars=['Country'],var_name="year", value_name="bmi")
df_melted
```

```python
df_melted.year = df_melted.year.map(lambda x: int(x[1:]))
```

```python
df_melted
```

```python
df_pivot = df_melted.pivot(index='Country',columns='year',values='bmi')
df_pivot
```
