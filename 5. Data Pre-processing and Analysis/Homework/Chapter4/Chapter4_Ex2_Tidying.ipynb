## Cau 2

```python
df1= pd.read_csv('input_data/billboard.csv',encoding='mac_latin2')
df1 
```

```python
df1_after = pd.melt(frame=df1, id_vars=['year','artist.inverted','track','time','genre','date.entered','date.peaked']
                   , var_name='week', value_name='rank')
df1_after
```

```python
df1_after['week'] = df1_after['week'].str.extract('(\d+)', expand=False).astype(int)
```

```python
df1_after =df1_after.dropna()
```

```python
df1_after['rank'] = df1_after['rank'].astype(int)
```

```python
df1_after
```

```python
# => nhiều đơn vị quan sát (Song & rank) trong 1 bản duy nhất

# Tạo một dataframe songs và các thông tin của nó
# Tạo một dataframe ranks chứa id của song và week kèm theo rank
```

```python
songs_cols = ["year", "artist.inverted", "track", "time", "genre"] 
songs = df1_after[songs_cols].drop_duplicates()
songs.reset_index(drop=True)
songs["song_id"] = songs.index
songs.head (10)
```

```python
ranks = pd.merge(df1_after, songs, on=["year", "artist.inverted", "track", "time", "genre"])
ranks = ranks [["song_id", "date.entered","date.peaked", "week", "rank"]]
ranks.head (10)
```

```python
# Tạo cột date 
ranks ['date'] = pd.to_datetime (ranks ['date.entered']) + pd.to_timedelta(ranks ['week'],unit='w')- pd.DateOffset(weeks=1)
```

```python
ranks.head()
```

```python
ranks = ranks.drop(['date.entered','date.peaked'], axis=1)
```

### Câu 3

```python
df2= pd.read_csv('input_data/tb-raw.csv')
df2
```

```python
# Problem
# Tên cột chứa giá trị thay vì chứa biến, ngoài ra nó còn chứa 2 thông tin là giới tính và độ tuổi 
# Trong cell chứa hỗn hợp giá trị (value, 0 và NaN)
```

```python
# Chuyển dữ liệu theo định dạng country, year, và sex_and_age
df2_after= pd.melt (df2, id_vars=["country", "year"],
value_name="cases", var_name="sex_and_age")
df2_after.head()
```

```python
# Tách sex_and_age thành 2 cột: sex và age
# Extract Sex, Age Lower bound and Age upper bound group 
tmp_df = df2_after["sex_and_age"].str.extract("(\D)(\d+)(\d{2})")

# Name columns
tmp_df.columns = ["sex", "age_lower","age_upper"]
# Create age`column based on "age_lower" and "age_upper"
tmp_df["age"] = tmp_df["age_lower"] + "-" + tmp_df["age_upper"]
```

```python
# Merge
df2_after = pd.concat([df2_after, tmp_df], axis=1)
df2_after.head ()
```

### Câu 4

```python
df3= pd.read_csv('input_data/weather-raw.csv')
df3
```

Vấn đề cần khắc phục
* Các biến được lưu trữ trong cả các dòng (tmin, tmax) và các cột (days).
* Tên cột chứa giá trị day thay vì chứa biến
* Trong từng cell chứa temperature nhưng có rất nhiều giá trị NaN
=> Để làm cho tập dữ liệu này gọn gàng | => di chuyển ba biến bị đặt sai (tmin, tmax và day) thành ba cột riêng lẻ: tmin. tmax và date.

```python
df3_after = pd.melt (df3, id_vars=["id", "year", "month", "element"], var_name="day_raw")
df3_after.head (10)
```

```python
# Tạo cột day chỉ chứa ngày
df3_after["day"] = df3_after["day_raw"].str.extract("d(\d+)", expand=False)
```

```python
df3_after
```

```python
# Chuyển dữ liệu thành số 
df3_after[["year", "month", "day"]] = df3_after[["year", "month","day"]].apply(lambda x: pd.to_numeric(x, errors='ignore'))
```

```python
df3_after
```

```python
# Tạo cột date chứa cả day, month, year 
import datetime
def create_date_from_year_month_day (row): 
    return datetime.datetime (year=row["year"],month=int(row["month"]), day=row["day"])
```

```python
df3_after["date"]= df3_after.apply(lambda row: create_date_from_year_month_day(row),axis=1)
```

```python
# bỏ các cột thừa
df3_after = df3_after.drop(['year', "month","day", "day_raw"], axis=1)
```

```python
# Bỏ các dòng dữ liệu Nan
df3_after = df3_after.dropna ()
```
