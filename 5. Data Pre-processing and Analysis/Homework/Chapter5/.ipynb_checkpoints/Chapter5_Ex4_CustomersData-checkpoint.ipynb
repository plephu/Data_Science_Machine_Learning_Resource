```python
# Phân tích đơn biến: trực quan hóa, kiểm tra dữ liệu outlier 
# Trực quan hóa dữ liệu cho từng biến Liên tục 
plt.figure(figsize=(8,5))
plt.subplot (1,2,1) 
sns.boxplot (df.income, orient='v')
plt.subplot (1,2,2)
sns.boxplot(df.age, orient='v')
plt.show()

# => Cả hai biến liên tục income và age đều không có outlier
```

```python
# Dataset có cột gender: Là thuộc tính phân loại kiểu chuỗi => tạo cột phân loại kiểu số
df = pd.get_dummies (df, drop_first=True)
```

* Customer Segmentation là một bài toán phân nhóm dựa trên sự tương tự về các thuộc tính và sẽ tính khoảng cách để biết mẫu này “gần” mẫu kia hay không bằng công thức tính khoảng cách Euclide
* Đối với các thuật toán cần tính khoảng cách thì dữ liệu trên các cột phải đảm bảo tính công bằng khi tính toán. Tuy nhiên, quan sát thấy income và age có range dữ liệu khác biệt, chênh lệch nhiều => cần phải chuẩn hóa dữ liệu.
* Muốn chuẩn hóa dữ liệu: cần kiểm tra phân phối của dữ liệu.

```python
plt.figure(figsize=(10,5)) 
plt.subplot(1, 2, 1)
sns.distplot(df.income) 
plt.subplot(1, 2, 2) 
sns.distplot(df.age)
plt.show()
```

Từ những kết quả trên ta thấy:
* Dữ liệu không theo phân phối Gaussian 
* Dữ liệu không có outlier
 * => Dùng MinMaxScaler để chuẩn hóa

### MinMaxScaler

```python
mmScaler = MinMaxScaler()
mmScaler.fit(df[['income','age']])
df_sub = mmScaler.transform(df[['income','age']])
```

```python
df_sub
```

```python
data_sub_min_max_scaler = pd.DataFrame(df_sub,columns=['mm_income', 'mm_age'])
df = pd.concat([df.reset_index(drop=True), data_sub_min_max_scaler], axis=1)
```

```python
df
```

```python
df.info()
```

```python
# Income & Age: trước khi Scaler
plt.figure(figsize=(8,8))
sns.scatterplot(data=df, x='income', y='age')
plt.show()
```

```python
# Income & Age: sau khi Scaler giá trị nằm trong khoảng [0,1]
plt.figure(figsize=(8,8))
sns.scatterplot(data=df, x='mm_income', y='mm_age')
plt.show()
```
