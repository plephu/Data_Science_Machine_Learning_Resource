```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### input category ordinal

```python
house_area = pd.DataFrame({"Area": [21, 32, 64, 56, 128, 160],
                          "Type": ["Small", "Small", "Medium", 
                                   "Medium", "Large", "Large"]})
```

```python
house_area
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
encoded_data = encoder.fit_transform(house_area['Type'])
```

```python
house_area['encoded_type'] = encoded_data
house_area
```

```python
sort_type =[]
for t in house_area.Type:
    if t == 'Small':
        sort_type.append(f'0_{t}')
    if t == 'Medium':
        sort_type.append(f'1_{t}')    
    if t == 'Large':
        sort_type.append(f'2_{t}')    
```

```python
house_area['sort_type'] = sort_type
house_area
```

```python
encoded_data = encoder.fit_transform(house_area['sort_type'])
```

```python
house_area['encoded_sort_type'] =encoded_data
house_area
```

###  Onehot Encoder

```python
students = pd.DataFrame({"Name": ["John", "Lucy", "Lyly", "Mark"],
                        "Sex": ["Male", "Female", "Other", "Male"],
                        "Age": [16, 17, 16, 17]})
```

```python
students
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(drop='first')
```

```python
encode_data = encoder.fit_transform(students[['Sex']]).toarray()
encode_data
```

```python
encoder.categories_[0]
```

```python
encoded_df = pd.DataFrame(encode_data, columns=encoder.categories_[0][1:]).add_prefix('E_')
```

```python
pd.concat([students, encoded_df], axis=1)
```

### Dummy encoder

```python
pd.get_dummies(students, columns=['Sex'], drop_first=True, prefix ='E')
```

### Handle Uncommon Category

```python
df = pd.read_csv('./input_data/Combined_DS_v10.csv')
df
```

```python
count_country = df.Country.value_counts()
count_country
```

```python
count_country[count_country < 10].index
```

```python
index_rare = df[df.Country.isin(count_country[count_country<10].index)].index
index_rare
```

```python
df.loc[index_rare,'Country'] = 'Rare'
```

```python
df.Country.value_counts()
```

### Binarizer values

```python
df['Paid_job'] = 0
```

```python
df.loc[df.ConvertedSalary > 0, 'Paid_job'] = 1
```

```python
df.head()
```

```python
df['Paid_job_lambda'] = df.ConvertedSalary.apply(lambda x: 1 if x > 0 else 0)
df.head()
```

```python
from sklearn.preprocessing import Binarizer
```

```python
binarizer = Binarizer(threshold=1)
```

```python
df['Paid_job_sklearn'] = binarizer.fit_transform(df[['ConvertedSalary']].fillna(0))
```

```python
df.head()
```

### Binning values

```python
df.head()
```

```python
pd.cut(df.ConvertedSalary,bins=5)
```

```python
bins_limit = [-np.inf, 10_000, 50_000, 100_000, 150_000, np.inf]
bins_label = ['Very low', 'Low', 'Medium', 'High','Very High']
```

```python
df['Binning_salary'] = pd.cut(df.ConvertedSalary, bins=bins_limit, labels=bins_label)
```
