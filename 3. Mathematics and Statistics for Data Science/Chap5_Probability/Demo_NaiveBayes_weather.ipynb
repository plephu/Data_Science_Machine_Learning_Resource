# Weather & Play

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('data/weather.xls', index_col=0)
data.info()
```

```python
data.head()
```

```python
# step 1: tạo bảng crosstab có index là Weather, column là Play
data_crosstab = pd.crosstab(data['Weather'],
                            data['Play'], 
                            margins = False)
data_crosstab
```

```python
data_crosstab['Total'] = data_crosstab['No'] + data_crosstab['Yes']
data_crosstab
```

```python
# Step 2    
# tính xác suất từng Weather
data_crosstab["P_Weather"] = data_crosstab["Total"] / data.shape[0]
```

```python
# tính xác suất Play = No
Total_No = sum(data_crosstab["No"].values)
P_No = Total_No/data.shape[0]
P_No
```

```python
# tính xác suất Play = Yes
Total_Yes = sum(data_crosstab["Yes"].values)
P_Yes = Total_Yes/data.shape[0]
P_Yes
```

```python
# tính P(Weather|Play)
data_crosstab["P_Weather_No"] = data_crosstab["No"]/Total_No
data_crosstab["P_Weather_Yes"] = data_crosstab["Yes"]/Total_Yes
```

```python
data_crosstab
```

```python
sum(data_crosstab["Yes"].values)/sum(data_crosstab["Total"].values)
```

```python
# Step 3
def predict_target(data, weather):
    # lấy dòng có Weather cần tìm
    event = data.loc[weather,:]
    
    # P(Yes|Weather) = P(Weather|Yes) * P(Yes) / P(Weather)
    P_Yes_event = event['P_Weather_Yes'] * P_Yes / event['P_Weather']
    print(event)
    return P_Yes_event
```

```python
P_Yes_Sunny = predict_target(data_crosstab, "Sunny")
P_Yes_Sunny
```
