```python
A=df[['Length','Width','Height']]
A.head()
```

```python
# Decomposition
U, s, VT = np.linalg.svd(A)
```

```python
U
```

```python
s
```

```python
VT
```

```python
Sigma = np.zeros((A.shape[0],A.shape[1]))
Sigma[:A.shape[1],:A.shape[1]]= np.diag(s)
Sigma
```

```python
n_elements = 1
Sigma = Sigma[:,:n_elements]
VT = VT[:n_elements,:]
```

```python
# Reconstruct
B=U.dot(Sigma.dot(VT))
B
```

```python
# Transform
T = U.dot(Sigma)
T
```

```python
T = A.dot(VT.T)
T
```

## 1.3

```python
A=df[['Length','Width','Height']]
A.head()
```

```python
#step 1
mu =np.mean(A)
X_hat = A- mu
X_hat.head()
```

```python
#step 2
S=np.cov(X_hat,rowvar=False)

eig_val, eig_vec = np.linalg.eig(S)
print('EigenValue: ',eig_val)
print('EigenVectors: \n',eig_vec)
```

```python
eig_val/sum(eig_val)
```

```python
#step 3
np.cumsum(eig_val)/sum(eig_val)
```

```python
plt.plot(range(1,4),eig_val)
plt.title('Scree plot')
plt.show()
```

```python
#step 4
k =2
B = eig_vec[:,:k]
PC= X_hat.dot(B)
PC.head()
```

```python
PC.columns = ['Comp1','Comp2']
PC['Gender'] = df['Gender']
PC.head()
```

## 1.4

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data = PC,x = PC['Comp1'], y = PC['Comp2'], hue ='Gender')
```

```python
def Distance(A, v):
    A_v = A - v
    distances = []
    for i in range(A_v.shape[0]):
        distances.append(norm(A_v[i,:])) # norm 2
    result = np.array(distances) 
    return result
```

```python
def KNN(A, v, k=3):
    distance = Distance(A, v)
    knn=pd.DataFrame(distance, columns =['distance']).sort_values(by='distance').head(k)
    return knn.index, knn.values
```

```python
A=np.array(df[['Length','Width','Height']])
A
```

```python
v =np.array([129,95,46])
v
```

```python
ind,dis = KNN(A, v, 3)
```
