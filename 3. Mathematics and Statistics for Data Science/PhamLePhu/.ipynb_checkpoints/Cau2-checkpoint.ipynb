### Chọn column Population_Driver_licence(%) 

```python
X = df['Population_Driver_licence(%)'].values
y = df['Petrol_Consumption'].values
```

```python
def gradient_descent(alpha, X, y, numIterations):
    m = X.shape[0] 
    n = Xbar.shape[1]
    w = np.ones(n)
    
    for i in range(numIterations):
        error = Xbar.dot(w) - y
        diff_w0 = sum(error) / m
        diff_w1 = sum(error * X[:,1]) / m
        J = np.sum(error ** 2) / (2 * m)
        
        gradient = np.array([diff_w0, diff_w1])
        w = w - alpha * gradient
        
    return w
```

```python
m = X.shape[0] 
Xbar = np.c_[np.ones(m), X] 
w = gradient_descent(0.1, Xbar, y, 100000)
w
```

```python
# Dùng ma trận giả đảo (X.w = y)
m = X.shape[0]
Xbar = np.c_[np.ones(m), X]
w = pinv(Xbar).dot(y)
w
```

## 2.3

```python
for i in range(X.shape[0]):
    y_predict = w[0] + w[1]*X
    plt.scatter(X, y, marker='o')
    plt.plot(X, y_predict, c='r')
    plt.xlabel('Population Driver licence(%)')
    plt.ylabel('Petrol Consumption')
```

```python
# Gradient Descent và bằng ma trận giả đảo cho kết quả gần giống nhau.
# Population Driver licence(%) và Petrol Consumption có mối quan hệ tuyến tính
```
