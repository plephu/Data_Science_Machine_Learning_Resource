## Yêu cầu 2: Tìm norm của vector
- Tạo vector v có các phần tử: 1, 3, 2, 4
- Tính norm 1, norm 2 và max norm của vector v

```python
# norm
from numpy.linalg import norm
v = np.array([1,3,2,4])
```

```python
# norm 1
norm(v,1)
```

```python
# norm 2
norm(v)
```

```python
# max norm

import math
norm(v,math.inf)
```

## Yêu cầu 3: Tìm norm 2 các vector dòng trong ma trận
- Đọc tập tin iris.csv ra dataframe df
- Từ dataframe df hãy tạo ma trận A với các thuộc tính 'sepal_length','sepal_width','petal_length','petal_width'
- Tính norm 2 cho mỗi vector dòng của ma trận A

```python
import pandas as pd
```

```python
df = pd.read_csv('data/iris.csv')
```
