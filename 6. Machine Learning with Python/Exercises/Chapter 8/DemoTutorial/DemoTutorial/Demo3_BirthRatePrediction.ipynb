```python
import pandas as pd
```

## 1. Reading some data

```python
data = pd.read_csv('data/TeenBirths.txt',delimiter='\t')
```

```python
data.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data['Location']=le.fit_transform(data['Location'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,:-1], data.iloc[:,-1], test_size = 0.3, random_state = 0)
```

## 2. Let's try some boosting models for regression

```python
# build the lightgbm model
from sklearn.metrics import mean_squared_error as mse
import lightgbm as lgb
clf = lgb.LGBMRegressor()
clf.fit(X_train,y_train)
preds = clf.predict(X_test)
print(mse(y_test,preds))
```

```python
from catboost import CatBoostRegressor
cb = CatBoostRegressor()
cb.fit(X_train,y_train)
preds = cb.predict(X_test)
print(mse(y_test,preds))
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(X_train,y_train)
preds = xgb.predict(X_test)
print(mse(y_test,preds))
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada = AdaBoostRegressor()
ada.fit(X_train,y_train)
preds = ada.predict(X_test)
print(mse(y_test,preds))
```
