## 1. Read data

```python
data = pd.read_csv("dataset_group.csv",header=None,names=['Date','TID','item'])
```

```python
data.head()
```

```python
data_new = data.groupby(['TID'])['item'].apply(list)
data_new = data_new.index.to_frame(name='TID').join(data_new)
data_new = data_new.reset_index(drop=True)
```

```python
data_new
```

```python
data_final = pd.DataFrame(data_new.item.values.tolist(), data_new.index)
```

```python
data_final
```

```python
from pyECLAT import ECLAT
eclat_instance = ECLAT(data=data_final, verbose=True)
```

```python
eclat_instance.df_bin
```

## find top 10 bought item?

```python
items_total = eclat_instance.df_bin.astype(int).sum(axis=0)
sorted_item = items_total.sort_values(ascending=False)
```

```python
sorted_item[:10]
```

```python
eclat_instance.uniq_
```

```python
df = pd.DataFrame({'items': items_total.index, 'transactions': items_total.values}) 
# cloning pandas DataFrame for visualization purpose  
df_table = df.sort_values("transactions", ascending=False)
#  Top 5 most popular products/items
df_table.head(5).style.background_gradient(cmap='Blues')
```

### Using ECLAT

```python
# the item shoud appear at least at 5% of transactions
min_support = 10/100
# start from transactions containing at least 2 items
min_combination = 2
# up to maximum items per transaction
max_combination = 3
rule_indices, rule_supports = eclat_instance.fit(min_support=min_support,
                                                 min_combination=min_combination,
                                                 max_combination=max_combination,
                                                 separator=' & ',
                                                 verbose=True)
```

```python
result = pd.DataFrame(rule_supports.items(),columns=['Item', 'Support'])
result.sort_values(by=['Support'], ascending=False)
```

### Mix it with MLXtend Association Rules

```python
from mlxtend.frequent_patterns import association_rules
```

```python
result = result.rename(columns={"Item": "itemsets", "Support": "support"})
```

```python
result
```

```python
for row in result['itemsets']:
    if ("milk" in row):
        print(row)
```

```python
new_column = []

for row in result['itemsets']:
    r = row.split('&')
    r = tuple(i.strip() for i in r)
    new_column.append(r)

result['itemsets'] = pd.Series(new_column)
```

```python
result
```

```python
rules_mlxtend = association_rules(result, support_only=True,metric="confidence", min_threshold=0.1)
print("number of rules: ",rules_mlxtend.shape[0])
rules_mlxtend.head()
```
