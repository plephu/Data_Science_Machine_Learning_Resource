```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('playgolf_data.xlsx',index_col=0)
df.head()
```

```python
features = df.drop('Play Golf', axis=1)
target = df[['Play Golf']]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
features = pd.get_dummies(features)
features
```

```python
from sklearn.utils.validation import column_or_1d
le = LabelEncoder()
target = le.fit_transform(column_or_1d(target))
```

```python
# Sử dụng BernoulliNB
from sklearn.naive_bayes import BernoulliNB
```

```python
model1 = BernoulliNB()
model1.fit(features, target)
```

```python
print('Accuracy is:', model1.score(features,target),'%')
```

```python
class_name = model1.classes_
class_name
```
