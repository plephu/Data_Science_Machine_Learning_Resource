## Question 1: Using Pipeline and Imputer to fill null

```python
from sklearn.pipeline import Pipeline
from sklearn import preprocessing
from sklearn.impute import SimpleImputer
pipeline_data = Pipeline(steps=[
    
    ("num_imputer", SimpleImputer(strategy="mean")),
     ('scaler', preprocessing.StandardScaler())
])
```

```python
data_new = pipeline_data.fit_transform(data)
```

```python
data_new = pd.DataFrame(data_new,columns=data.columns)
print(data_new.isna().sum())
```

```python
data_new.to_csv("scaled_customer_data.csv")
```

## Question 2: Select Clustering Model

```python
from sklearn.cluster import KMeans
X1 = data_new.values # dataframe -> array
inertia = []
for n in range(3 , 15):
    algorithm = (KMeans(n_clusters = n ,init='k-means++', n_init = 10 ,max_iter=500, 
                        tol=0.0001,  random_state= 111  , algorithm='elkan') )
    algorithm.fit(X1)
    inertia.append(algorithm.inertia_)
```

```python
import numpy as np
plt.figure(1 , figsize = (15 ,6))
plt.plot(np.arange(3 , 15) , inertia , 'o')
plt.plot(np.arange(3 , 15) , inertia , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
plt.show()
```

```python
y = inertia
print(len(y))

x = range(3, 15)
from kneed import KneeLocator
kn = KneeLocator(x, y, curve='convex', direction='decreasing')
print(kn.knee)


import matplotlib.pyplot as plt
plt.xlabel('number of clusters k')
plt.ylabel('Sum of squared distances')
plt.plot(x, y, 'bx-')
plt.vlines(kn.knee, plt.ylim()[0], plt.ylim()[1], linestyles='dashed')
```

### Như vậy, dựa trên kết quả phân tích trên có thể chọn k=4 hoặc k=6 hoặc k=8

```python
from sklearn.mixture import GaussianMixture 
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
n_clusters=np.arange(3, 15)
sils=[]
sils_err=[]
iterations=20
for n in n_clusters:
    tmp_sil=[]
    gmm=GaussianMixture(n, n_init=2).fit(data_new) 
    labels=gmm.predict(data_new)
    silhouette_avg = silhouette_score(data_new, labels,metric='euclidean')
    sils.append(silhouette_avg)
    print("Clustering with n = {0} clusters, avg silhoutte score = {1}".format(n,silhouette_avg))


```

```python
import numpy as np
plt.figure(1 , figsize = (15 ,6))
plt.plot(np.arange(3 , 15) , sils , 'o')
plt.plot(np.arange(3 , 15) , sils , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('Silhouette score')
plt.show()
```

### Nhận xét: có thể chọn số cluster = 3 (silhoutte score cao nhất) trong GMM

```python
from sklearn.cluster import AgglomerativeClustering
n_clusters=np.arange(3, 15)
sils=[]
sils_err=[]
iterations=20
for n in n_clusters:
    tmp_sil=[]
    agg=AgglomerativeClustering(n).fit(data_new) 
    labels=agg.labels_
    silhouette_avg = silhouette_score(data_new, labels,metric='euclidean')
    sils.append(silhouette_avg)
    print("Clustering with n = {0} clusters, avg silhoutte score = {1}".format(n,silhouette_avg))
import numpy as np
plt.figure(1 , figsize = (15 ,6))
plt.plot(np.arange(3 , 15) , sils , 'o')
plt.plot(np.arange(3 , 15) , sils , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('Silhouette score')
plt.show()
```

### Nhận xét: có thể chọn số cluster = 3 (silhoutte score cao nhất) với AgglomerativeClustering

#### Nhận xét:
" each principal component is the linear combination of the Y variables that accounts for the greatest possible variance."
Qua kết quả PCA ta thấy Principal components 1 có hệ số cao ở các features liên quan tới Purchases (số lượng giao dịch / mua sắm) thể hiện mức độ mua sắm

Trong khi đó Pricipal component 2 có hệ số cao ở các feature về cash (tiền mặt), ví dụ cash_advance (lượng tiền mặt có trước), balance (tiền trong tài khoản) => PC2 liên quan tới mức độ tài sản và khả năng thanh toán

Từ kết quả clustering của hình Clusters wrt AgglomerativeClustering and PCA, với PC1 là trục x và PC2 là trục y ta có thể chia thành 3 nhóm người dùng:


+ nhóm 1 (pink): có mức mua sắm trung bình - cao
+ nhóm 2 (yellow): mức mua sắm trung bình - thấp (thấp hơn nhóm 1) nhưng có khả năng thanh toán hoặc tài sản cao hơn nhóm 1. Lưu ý nhóm 2 có vài outliers
+ nhóm 3 (màu xanh tím): mức mua sắn trung bình - thấp (thấp hơn nhóm 1) khả năng thanh toán hoặc tài sản thấp hơn nhóm 2
