## Question 1: Using Pipeline and Imputer to fill null

```python
from sklearn.pipeline import Pipeline
from sklearn import preprocessing
from sklearn.impute import SimpleImputer
pipeline_data = Pipeline(steps=[
    
    ("num_imputer", SimpleImputer(strategy="mean")),
     ('scaler', preprocessing.StandardScaler())
])
```

```python
data_new = pipeline_data.fit_transform(data)
```

```python
data_new = pd.DataFrame(data_new,columns=data.columns)
print(data_new.isna().sum())
```

```python
data_new.to_csv("scaled_customer_data.csv")
```

```python
data_new.BALANCE.mean()
```

## Question 2: Select Clustering Model

```python
from sklearn.cluster import KMeans
X1 = data_new.values # dataframe -> array
inertia = []
for n in range(3 , 15):
    algorithm = (KMeans(n_clusters = n ,init='k-means++', n_init = 10 ,max_iter=500, 
                        tol=0.0001,  random_state= 111  , algorithm='elkan') )
    algorithm.fit(X1)
    inertia.append(algorithm.inertia_)
```

```python
import numpy as np
plt.figure(1 , figsize = (15 ,6))
plt.plot(np.arange(3 , 15) , inertia , 'o')
plt.plot(np.arange(3 , 15) , inertia , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
plt.show()
```

```python
y = inertia
print(len(y))

x = range(3, 15)
from kneed import KneeLocator
kn = KneeLocator(x, y, curve='convex', direction='decreasing')
print(kn.knee)


import matplotlib.pyplot as plt
plt.xlabel('number of clusters k')
plt.ylabel('Sum of squared distances')
plt.plot(x, y, 'bx-')
plt.vlines(kn.knee, plt.ylim()[0], plt.ylim()[1], linestyles='dashed')
```

### NhÆ° váº­y, dá»±a trÃªn káº¿t quáº£ phÃ¢n tÃ­ch trÃªn cÃ³ thá»ƒ chá»n k=4 hoáº·c k=6 hoáº·c k=8

```python
from sklearn.mixture import GaussianMixture 
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
n_clusters=np.arange(3, 15)
sils=[]
sils_err=[]
iterations=20
for n in n_clusters:
    tmp_sil=[]
    gmm=GaussianMixture(n, n_init=2).fit(data_new) 
    labels=gmm.predict(data_new)
    silhouette_avg = silhouette_score(data_new, labels,metric='euclidean')
    sils.append(silhouette_avg)
    print("Clustering with n = {0} clusters, avg silhoutte score = {1}".format(n,silhouette_avg))


```

```python
import numpy as np
plt.figure(1 , figsize = (15 ,6))
plt.plot(np.arange(3 , 15) , sils , 'o')
plt.plot(np.arange(3 , 15) , sils , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('Silhouette score')
plt.show()
```

### Nháº­n xÃ©t: cÃ³ thá»ƒ chá»n sá»‘ cluster = 3 (silhoutte score cao nháº¥t) trong GMM

```python
from sklearn.cluster import AgglomerativeClustering
n_clusters=np.arange(3, 15)
sils=[]
sils_err=[]
iterations=20
for n in n_clusters:
    tmp_sil=[]
    agg=AgglomerativeClustering(n).fit(data_new) 
    labels=agg.labels_
    silhouette_avg = silhouette_score(data_new, labels,metric='euclidean')
    sils.append(silhouette_avg)
    print("Clustering with n = {0} clusters, avg silhoutte score = {1}".format(n,silhouette_avg))
import numpy as np
plt.figure(1 , figsize = (15 ,6))
plt.plot(np.arange(3 , 15) , sils , 'o')
plt.plot(np.arange(3 , 15) , sils , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('Silhouette score')
plt.show()
```

### Nháº­n xÃ©t: cÃ³ thá»ƒ chá»n sá»‘ cluster = 3 (silhoutte score cao nháº¥t) vá»›i AgglomerativeClustering

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_new)
```

```python
X1 = data_pca # dataframe -> array
inertia = []
algorithm = AgglomerativeClustering(n_clusters =3)
algorithm.fit(X1)
```
