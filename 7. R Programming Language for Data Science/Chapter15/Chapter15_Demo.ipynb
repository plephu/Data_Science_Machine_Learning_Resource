# Normal Distribution

## dnorm

```r
x =  seq(-10,10,by = .1)
```

```r
y = dnorm(x, mean = 0, sd = 1)
```

```r
plot(x,y)
```

```r
x_test = 2.5
y_test = dnorm(x_test, mean = 0, sd =1)
y_test
```

### pnorm

```r
x =  seq(-10,10,by = .1)
y = pnorm(x, mean = 0, sd = 1)
plot(x,y)
```

```r
x_test = 2.5
y_test = pnorm(x_test, mean = 0, sd =1)
y_test
```

## qnorm

```r
x =  seq(0,1,by = 0.02)
y = qnorm(x, mean = 0, sd = 1)
plot(x,y)
```

```r
x_test = 0.6
y_test = qnorm(x_test, mean = 0, sd = 1)
y_test
```

## rnorm

```r
y = rnorm(50)
y
```

```r
hist(y, main = 'Normal Distribution')
```

# Binomial Distribution

## dbinom

```r
x = seq(0,50, by=1)
y= dbinom(x,50,0.5)
plot(x,y)
```

```r
x_test = 30
y_test = dbinom(x_test,50,0.5)
y_test
```

```r
p1 = dbinom(2,3,0.6)
p1
```

## pbinom

```r
p_x_25_less = pbinom(25,50,0.5)
p_x_25_less
```

```r
p_x_25_more = 1- pbinom(24,50,0.5)
p_x_25_more
```

## qbinom

```r
x= qbinom(0.25,50,0.5)
x
```

## rbinom

```r
x= rbinom(8,100,1/6)
x
```
