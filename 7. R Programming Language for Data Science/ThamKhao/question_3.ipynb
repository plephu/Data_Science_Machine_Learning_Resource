## Câu 1
Sử dụng Linear Regression để thực hiện việc dự đoán sales dựa trên
thuộc tính youtube

```r
data = read.csv('data/marketing.csv')
```

```r
head(data)
```

```r
str(data)
```

```r
summary(data)
```

```r
library(ggplot2)
```

```r
ggplot(data, aes(x = youtube, y = sales)) +
  geom_point() +
  labs(title = "Moi lien he giua 'sales' va 'youtube'",
       x = "Youtube",
       y = "Sales") +
  theme_minimal()
```

```r
# Kiểm tra dữ liệu thiếu
any(is.na(data))
```

```r
# Kiểm tra và xử lý outliers cho cột 'sales'
sales_median <- median(data$sales)
MAD <- median(abs(data$sales - sales_median))
outlier_threshold <- 3 * MAD
data$sales[data$sales - sales_median > outlier_threshold] <- sales_median + outlier_threshold
data$sales[sales_median - data$sales > outlier_threshold] <- sales_median - outlier_threshold
```

```r
# Đặt seed để đảm bảo tái lặp kết quả
set.seed(123)

# Tạo một vector với các chỉ số ngẫu nhiên từ 1 đến số lượng dòng của data
sample_indices <- sample(1:nrow(data), nrow(data))

# Tính toán số lượng dòng cho tập huấn luyện và tập kiểm tra
split_ratio <- 0.7 
split_index <- floor(nrow(data) * split_ratio)

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
train_data <- data[sample_indices[1:split_index], ]
test_data <- data[sample_indices[(split_index + 1):nrow(data)], ]
```

```r
# Xây dựng mô hình hồi quy tuyến tính trên train_data
model <- lm(sales ~ youtube, data = train_data)

# Xem kết quả mô hình
summary(model)
```

```r
# Dự đoán giá trị 'sales' trên tập kiểm tra
y_test_pred <- predict(model, newdata = test_data)

# Lấy giá trị thực tế của 'sales' từ tập kiểm tra
y_test <- test_data$sales

# So sánh giá trị dự đoán và giá trị thực tế
comparison <- data.frame(ThucTe = y_test, DuDoan = y_test_pred)

# In ra và xem kết quả
print(comparison)
```

```r
# Tính MSE cho tập huấn luyện
y_train <- train_data$sales
mse_train <- mean((y_train - predict(model, newdata = train_data))^2)

# Tính MSE cho tập kiểm tra
mse_test <- mean((y_test - y_test_pred)^2)

# In ra MSE cho tập huấn luyện và tập kiểm tra
cat("MSE trên tập huấn luyện:", mse_train, "\n")
cat("MSE trên tập kiểm tra:", mse_test, "\n")
```

```r
# Tính R² cho tập huấn luyện
ssr_train <- sum((y_train - predict(model, newdata = train_data))^2)
sst_train <- sum((y_train - mean(y_train))^2)
r_squared_train <- 1 - (ssr_train / sst_train)

# Tính R² cho tập kiểm tra
ssr_test <- sum((y_test - y_test_pred)^2)
sst_test <- sum((y_test - mean(y_test))^2)
r_squared_test <- 1 - (ssr_test / sst_test)

# In ra R² cho tập huấn luyện và tập kiểm tra
cat("R² trên tập huấn luyện:", r_squared_train, "\n")
cat("R² trên tập kiểm tra:", r_squared_test, "\n")
```

Nhận xét:
- Mô hình ko bị overfitting

```r
# Lấy hệ số của các biến độc lập
coefficients <- coef(model)

# Lấy hệ số chặn (intercept)
intercept <- coef(model)[1]

# In ra hệ số của các biến độc lập
cat("Hệ số của các biến độc lập:\n")
print(coefficients[-1])  # Bỏ qua hệ số chặn

# In ra hệ số chặn
cat("Hệ số chặn (intercept):\n")
print(intercept)
```

```r
# Tạo DataFrame chứa giá trị 'youtube' cho dự đoán
values_to_predict <- data.frame(
  youtube = c(100, 200, 300))

# Dự đoán 'sales' cho các giá trị
sales_predictions <- predict(model, newdata = values_to_predict)

# In ra kết quả dự đoán
print(sales_predictions)
```

```r
# Xây dựng mô hình hồi quy tuyến tính trên train_data
model <- lm(sales ~ youtube + facebook + newspaper, data = train_data)

# Xem kết quả mô hình
summary(model)
```

```r
# Dự đoán giá trị 'sales' trên tập kiểm tra
y_test_pred <- predict(model, newdata = test_data)

# Lấy giá trị thực tế của 'sales' từ tập kiểm tra
y_test <- test_data$sales

# So sánh giá trị dự đoán và giá trị thực tế
comparison <- data.frame(ThucTe = y_test, DuDoan = y_test_pred)

# In ra và xem kết quả
print(comparison)
```

```r
# Tính MSE cho tập huấn luyện
y_train <- train_data$sales
mse_train <- mean((y_train - predict(model, newdata = train_data))^2)

# Tính MSE cho tập kiểm tra
mse_test <- mean((y_test - y_test_pred)^2)

# In ra MSE cho tập huấn luyện và tập kiểm tra
cat("MSE trên tập huấn luyện:", mse_train, "\n")
cat("MSE trên tập kiểm tra:", mse_test, "\n")
```

```r
# Tính R² cho tập huấn luyện
ssr_train <- sum((y_train - predict(model, newdata = train_data))^2)
sst_train <- sum((y_train - mean(y_train))^2)
r_squared_train <- 1 - (ssr_train / sst_train)

# Tính R² cho tập kiểm tra
ssr_test <- sum((y_test - y_test_pred)^2)
sst_test <- sum((y_test - mean(y_test))^2)
r_squared_test <- 1 - (ssr_test / sst_test)

# In ra R² cho tập huấn luyện và tập kiểm tra
cat("R² trên tập huấn luyện:", r_squared_train, "\n")
cat("R² trên tập kiểm tra:", r_squared_test, "\n")
```

Nhận xét:
- Mô hình ko bị overfitting và có khả năng dự đoán tốt

```r
# Lấy hệ số của các biến độc lập
coefficients <- coef(model)

# Lấy hệ số chặn (intercept)
intercept <- coef(model)[1]

# In ra hệ số của các biến độc lập
cat("Hệ số của các biến độc lập:\n")
print(coefficients[-1])  # Bỏ qua hệ số chặn

# In ra hệ số chặn
cat("Hệ số chặn (intercept):\n")
print(intercept)
```

```r
# Tạo DataFrame chứa giá trị 'youtube', 'facebook', 'newspaper' cho dự đoán
values_to_predict <- data.frame(
  youtube = c(100, 200, 300),
  facebook = c(50, 75, 125), # Thay thế bằng giá trị 'facebook' tương ứng
  newspaper = c(30, 40, 60) # Thay thế bằng giá trị 'newspaper' tương ứng
)

# Dự đoán 'sales' cho các giá trị
sales_predictions <- predict(model, newdata = values_to_predict)

# In ra kết quả dự đoán
print(sales_predictions)
```
