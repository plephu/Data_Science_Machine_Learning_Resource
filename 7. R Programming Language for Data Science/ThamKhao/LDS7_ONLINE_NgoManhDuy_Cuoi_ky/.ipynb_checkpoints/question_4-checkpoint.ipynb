```r
# xoa cot X va veil.type (chi co 1 factor)
data <- subset(data, select = -c(X, veil.type))
data
```

### 4. Tạo train và test từ dữ liệu data.

```r
set.seed(42)
trainingRowIndex <- sample(1:nrow(data), 0.7*nrow(data))
print("Selected training row indexes: ")
print(trainingRowIndex)
trainingData <- data[trainingRowIndex,]
testData <- data[-trainingRowIndex, ]
cat('Row in training data: ',nrow(trainingData))
cat('\nRow in test data: ',nrow(testData))
```

### Sử dụng Logistic Regression

### 5. Xây dựng model với train
### 6. In summary của model.

```r
mylogit <- glm(Class ~ ., data = trainingData, family='binomial')
print(summary(mylogit))
```

#### 7. Dự đoán y_pred từ test => so sánh với y_test.

```r
pred <- predict(mylogit,
                newdata = testData,
                type = 'response')
pred_value <- ifelse(pred > 0.5, 'Poisonous', 'Edible')
print('Testdata class vs predict')
result <- data.frame(testClass = testData$Class[30:40],pred_value[30:40])
print(result)
```

#### 8. Đánh giá model.

```r
misClasificError = mean(pred_value != testData$Class)
print(paste('Accuracy: ', 1 - misClasificError))
```

#### Nhận xét: xác suất dự đoán là 100%, có thể vì có nhiều thông tin nên việc dự đoán cho tỉ lệ chính xác cao.

#### 9. Trực quan hóa model.

```r
confusionMatrix(as.factor(pred_value),as.factor(testData$Class))
```
