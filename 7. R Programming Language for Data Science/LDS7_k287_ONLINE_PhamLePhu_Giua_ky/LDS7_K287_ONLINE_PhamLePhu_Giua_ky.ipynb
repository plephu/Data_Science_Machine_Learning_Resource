## Câu 1: Vectors, Factors, Matrices, Arrays, và Lists (1.0 điểm)
### Yêu cầu : Hãy thực hiện các yêu cầu được liệt kê bên dưới, có in kết quả của từng câu:

#### 1. Tạo vector có tên là heights chứa chiều cao (centimet) của 10 cá nhân: 175, 162, 183, 170, 158, 168, 172, 159, 165, 157. In heights.

```r
heights <- c(175, 162, 183, 170, 158, 168, 172, 159, 165, 157)
print(heights)
```

#### 2. Tạo vector có tên là height_type từ vector heights ở Câu 1 với giá trị của từng phần tử như sau: nếu <160 : "Short", >=160 & <=170: "Average" , >170 : "Tall". Chuyển vector này thành factor height_type. In factor height_type.

```r
height_type <- ifelse(heights < 160, "Short", ifelse(heights >= 160 & heights <= 170, "Average", "Tall"))
height_type <- factor(height_type, levels = c("Short", "Average", "Tall"))
print(height_type)
```

#### 3. Tạo matrix có tên student_scores (4,3) đại diện cho điểm của 4 học sinh trong 3 môn học, điểm là các giá trị ngẫu nhiên trong khoảng từ 7.0 đến 10.0. In student_scores.

```r
student_scores <- matrix(runif(4 * 3, min = 7.0, max = 10.0), nrow = 4, ncol = 3)
print(student_scores)
```

#### 4. Tính điểm trung bình cho mỗi học sinh trong ma trận student_scores từ Câu 3. In kết quả.

```r
average_scores <- rowMeans(student_scores)
print(average_scores)
```

#### 5. Tạo array có tên là sales_data (22, 12, 2) chứa doanh số bán hàng trong 1 năm của 2 sản phẩm; biết mỗi tháng ghi nhận doanh số 22 ngày, và 1 năm có 12 tháng, với các giá trị doanh thu ngẫu nhiên từ 1000 USD đến 5000 USD. In sales_data.

```r
sales_data <- array(runif(22 * 12 * 2, min = 1000, max = 5000), dim = c(22, 12, 2))
print(sales_data)
```

#### 6. Tính tổng doanh số bán hàng cho từng sản phẩm trong năm, sử dụng array sales_data từ câu trên. In kết quả.

```r
total_sales_per_product <- apply(sales_data, c(3), sum)
print(total_sales_per_product)
```

#### 7. Tạo một list có tên là data_collection chứa các thành phần sau: matrix student_scores từ Câu 3, array sales_data từ Câu 5, vector heights từ Câu 1 và factor height_type từ Câu 2. In data_collection.

```r
data_collection <- list(student_scores = student_scores,
                         sales_data = sales_data,
                         heights = heights,
                         height_type = height_type)

print(data_collection)
```

#### 8. Tính trung bình chiều cao trong list my_list từ Câu 7. In kết quả.

```r
heights <- data_collection$heights
average_height <- mean(heights)
print(average_height)
```

#### 9. Tính trung bình doanh thu cho mỗi sản phẩm trong năm trong list my_list ở câu 7. In kết quả.

```r
sales_data <- data_collection$sales_data
average_sales_per_product <- apply(sales_data, c(3), mean)
print(average_sales_per_product)
```

#### 10. Cho vector có tên là crazy_sale_dates <- c("2023-09-09", "2023-10-10", "2023-11-11", "2023-12-12"), hãy chuyển các phần tử chứa chuỗi ngày của vector này sang kiểu ngày (Date) sau đó đưa vector này vào cuối list data_collection ở câu 7. In crazy_sale_dates và list data_collection.

## Câu 2: Danh sách Sinh viên (Dataframe + Data Visualization) (1.0 điểm)
### Cho các tập tin dữ liệu liên quan đến sinh viên chứa trong folder **data**. Hãy thực hiện các yêu cầu được liệt kê bên dưới, có in kết quả của từng câu:

#### 1. Tạo dataframe df chứa danh sách các sinh viên khoa CN, khoa TO và khoa VL. Các danh sách này được đọc từ các tập tin sinh_vien_CN.csv, sinh_vien_TO.csv và sinh_vien_VL.csv trong folder data. Hiển thị thông tin chung của dữ liệu: head, tail, summary, số dòng, số cột của dữ liệu.


```r
library(tidyverse)
```

```r
data_folder <- "data/"

# Đọc dữ liệu từ các tập tin CSV và tạo các Data Frame cho mỗi khoa
df_cn <- read.csv(file.path(data_folder, "sinh_vien_CN.csv"))
df_to <- read.csv(file.path(data_folder, "sinh_vien_TO.csv"))
df_vl <- read.csv(file.path(data_folder, "sinh_vien_VL.csv"))

df <- rbind(df_cn, df_to, df_vl)

head(df)
tail(df)
summary(df)
print(paste("Số dòng, số cột:", nrow(df), ",", ncol(df), "\n"))
```

#### 2. Tạo dataframe df_ket_qua với nội dung được đọc ra từ tập tin ket_qua.csv trong folder data. Hiển thị thông tin chung của dữ liệu: head, tail, summary, số dòng, số cột của dữ liệu.

```r
df_ket_qua <- read.csv(file.path(data_folder, "ket_qua.csv"))
head(df_ket_qua)
tail(df_ket_qua)
summary(df_ket_qua)
print(paste("Số dòng, số cột:", nrow(df_ket_qua), ",", ncol(df_ket_qua)))
```

#### 3. Tạo dataframe df_dtb từ df_ket_qua bằng cách nhóm theo masv và tính điểm trung bình cho mỗi sinh viên (gợi ý: dùng group_by và summarize).

```r
df_dtb <- df_ket_qua %>%
  group_by(masv) %>%
  summarize(diem_trung_binh = mean(diem))

head(df_dtb)
tail(df_dtb)
summary(df_dtb)
```

#### 4. Nối df_dtb vào df theo masv để có một dataframe chứa cả thông tin sinh viên và điểm trung bình.

```r
df <- merge(df, df_dtb, by = "masv", all.x = TRUE)

head(df)
tail(df)
summary(df)
cat("Số dòng, số cột:", nrow(df), ",", ncol(df), "\n")
```

#### 5. Cho biết 5 sinh viên có điểm lớn nhất của mỗi khoa, sắp tăng theo mã khoa và giảm theo điểm.

```r
top_5_students <- df %>%
  group_by(makh) %>%
  top_n(5, diem_trung_binh) %>%
  arrange(makh, desc(diem_trung_binh))
top_5_students
summary(top_5_students)
```

#### 6. Trong dataframe df tạo thêm cột **ketqua** với kết quả là Giỏi nếu điểm>=8, kết quả là Khá nếu điểm<8 và điểm >=6.5, kết quả là Trung bình nếu điểm <6.5 và điểm >=5, kết quả là Yếu nếu điểm<5

```r
df_with_ketqua <- df %>%
  mutate(ketqua = case_when(
    diem_trung_binh >= 8 ~ "Giỏi",
    diem_trung_binh >= 6.5 ~ "Khá",
    diem_trung_binh >= 5 ~ "Trung bình",
    TRUE ~ "Yếu"
  ))

head(df_with_ketqua)
tail(df_with_ketqua)
summary(df_with_ketqua)
```

#### 7. Cho biết mỗi khoa có bao nhiêu sinh viên Khá? Vẽ barplot thể hiện số lượng sinh viên Khá của từng khoa.

```r
library(plotrix)
```

```r
count_kha <- df_with_ketqua %>%
  filter(ketqua == "Khá") %>%
  group_by(makh) %>%
  summarize(so_luong = n())

barplot(height = count_kha$so_luong, names.arg = count_kha$makh, main = "Số lượng sinh viên Khá của từng khoa",col = rainbow(length(count_kha$so_luong)))
```

#### 8. Vẽ pie chart cho cột gioitinh: 1 pie chart cho tất cả SV, và mỗi khoa một pie chart riêng. Nhận xét chung, nhận xét từng khoa.  

```r
# Tạo pie chart cho toàn bộ sinh viên
pie_all <- table(df$gioitinh)
pie_all # False: Nữ, True - Nam
names(pie_all) <- c("Nữ", "Nam")
pie(pie_all, main = "Phân bố giới tính của tất cả sinh viên")
```

- Nhận xét: Tổng số lượng sinh viên nữ ít hơn sinh viên nam, nhưng không chênh lệch nhiều.

```r
# Tạo pie chart cho từng khoa
pie_by_khoa <- df %>%
  group_by(makh) %>%
  summarize(nam = sum(gioitinh == 'True'), nu = sum(gioitinh == 'False'))
print(pie_by_khoa)
for (i in 1:nrow(pie_by_khoa)) {
  khoa_name <- pie_by_khoa$makh[i]
  pie_data <- c(pie_by_khoa$nu[i], pie_by_khoa$nam[i])
  names(pie_data) <- c("Nữ", "Nam")
  pie(pie_data, main = paste("Phân bố giới tính của khoa", khoa_name))
}
```

### Nhận xét
- Khoa CN, TO: Số lượng sinh viên nam và nữ gần tương đương, Nam hơn nữ 1 ít
- Khoa VL: Có sự chênh lệch đáng kể giữa sinh viên Nam và Nữ, Nam chiếm gần 60%

#### 9. Vẽ biểu đồ barplot thể hiện điểm trung bình của mỗi khoa. Nhận xét. (Gợi ý: tính điểm trung bình cho từng khoa và thể hiện lên biểu đồ)

```r
avg_scores <- df %>%
  group_by(makh) %>%
  summarize(avg_diem = mean(diem_trung_binh))

avg_scores

barplot(height = avg_scores$avg_diem, names.arg = avg_scores$makh, main = "Điểm trung bình của mỗi khoa", col = rainbow(length(avg_scores$avg_diem)))
```

- Nhận xét: điểm trung bình của 3 khoa gần bằng nhau, từ 7.1-7.3.
- Khoa TO có điểm TB cao nhất, Khoa VL thấp nhất.

#### 10. Vẽ biểu đồ thể hiện số lượng SV theo ketqua(Giỏi, Khá....) của mỗi khoa. Nhận xét. (mỗi khoa 1 biểu đồ)

```r
for (khoa in unique(df_with_ketqua$makh)) {
  khoa_data <- df_with_ketqua %>%
    filter(khoa == makh) %>%
    group_by(ketqua) %>%
    tally() %>%
    arrange(desc(n))
  barp(height = khoa_data$n, names.arg = khoa_data$ketqua, main = paste("Số lượng sinh viên theo kết quả của khoa", khoa), col = rainbow(length(khoa_data$n)))
  print(khoa_data)
}
```

## Nhận xét:
- Khoa CN: Sinh viên xếp loại khá chiếm đa số (hơn 80%), Sinh viên Trung bình và giỏi gần bằng nhau.
- Khoa TO: Sinh viên xếp loại khá chiếm đa số (hơn 80%), Sinh viên giỏi có số lượng gấp đôi sinh viên trung bình.
- Khoa VL: Sinh viên xếp loại khá chiếm đa số (hơn 80%), Sinh viên giỏi có số lượng thấp sinh viên trung bình.

```r
initial_data_types <- sapply(data, class)
columns_to_convert <- c("fruit_name", "fruit_subtype")
for (col in columns_to_convert) {
  if (initial_data_types[col] != "factor") {
    data[[col]] <- as.factor(data[[col]])
  }
}
sapply(data, class)
```

#### 9. Nhóm theo fruit_subtype, hãy thống kê số lượng mẫu theo từng subtype; max và min từng cột mass, width, height. (Gợi ý: dùng group_by() và summarize())

```r
result <- data %>%
  group_by(fruit_subtype) %>%
  summarize(
    count = n(),
    max_mass = max(mass),
    min_mass = min(mass),
    max_width = max(width),
    min_width = min(width),
    max_height = max(height),
    min_height = min(height)
  )
result
```
