```r
x = c(23, 45, 30, 29)
labels = c('Tin', 'Van', 'Toan', 'Ly')
pie(x, labels)
```

```r
pie(x, labels, main='Bieu do', col = rainbow(length(x)), clockwise = TRUE)
```

```r
pie_percent = round(x * 100/sum(x), 1)
pie(x, labels = paste(labels, pie_percent, '%'), 
    main='Bieu do phan tram', 
    col = rainbow(length(x)), 
    clockwise = TRUE)
legend('topright', labels, cex=0.8, fill=rainbow(length(x)))
```
