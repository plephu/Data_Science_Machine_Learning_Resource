```r
ggplot(ageinc, aes(y = income)) +
  geom_boxplot() +
  labs(y = "Income") +
  ggtitle("Income boxplot")
```

```r
ggplot(ageinc, aes(y = age)) +
  geom_boxplot() +
  labs(y = "Age") +
  ggtitle("Age boxplot")
```

```r
# Compute Hopkins statistic for ageinc 
res <- get_clust_tendency(ageinc[, 1:2], n = nrow(ageinc[, 1:2])-1, graph = TRUE)
res$hopkins_stat
```

```r
res$plot
```

```r
# 5. Xây dựng model từ dữ liệu income và age. 
```

```r
wss =(nrow(ageinc)-1) * sum(apply(ageinc[,1:2],2,var))
```

```r
wss
```

```r
for (i in 1:6) wss[i] = sum(kmeans(ageinc[,1:2], centers = i, nstart = 20)$withinss)
```

```r
print(wss)
```

```r
plot(1:6, wss, type='b', xlab = 'Number of cluster', ylab = 'WSS')
```

```r
# 6. Tìm kết quả => có bao nhiêu cụm => mẫu nào thuộc cụm nào? 
```

```r
# k=3
```

```r
ageincCluster = kmeans(ageinc[,1:2], centers = 3, nstart = 20)
```

```r
ageincCluster
```

```r
print(ageincCluster$center)
```

```r
# 7. Vẽ hình (với mỗi cụm là một màu) => xem kết quả. 
```

```r
library(factoextra)
```

```r
fviz_cluster(ageincCluster, geom = 'point', data =  ageinc[,1:2]) + ggtitle('k=3')
```

```r
fviz_cluster(ageincCluster, geom = 'point', data =  ageinc[,1:2], stand = F) + ggtitle('k=3')
```
