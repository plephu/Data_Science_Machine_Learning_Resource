# Yêu cầu 1

```r
# c. Vẽ biểu đồ quan sát mối liên hệ giữa sales và youtube 
```

```r
input <- data[,c('sales','youtube')]
head(input)
```

```r
scatter.smooth(x=input$youtube, y = input$sales, main='sales vs youtube')
```

```r
# d. Tiền xử lý dữ liệu 
```

```r
any(is.na(data))
```

```r
# e. Kiểm tra và xử lý outliers 
```

```r
par(mfrow = c(1,2))
boxplot(input$sales, main = 'sales', sub = paste('Outlier rows:', boxplot.stats(input$sales)$out))
boxplot(input$youtube, main = 'youtube', sub = paste('Outlier rows:', boxplot.stats(input$youtube)$out))
```

```r
# Không có outlier
```

```r
print(cor(input$sales, input$youtube))
```

```r
# f. Tạo train:test từ dữ liệu data với tỉ lệ 70:30 hoặc 80:20 
```

```r
set.seed(42)
trainingRowIndex <- sample(1:nrow(input),  0.7*nrow(input))
print("Selected training row index:")
print(trainingRowIndex)
trainingData <- input[trainingRowIndex,]
testData <- input[-trainingRowIndex,]
print("Rows of training data and test data:")
print(nrow(trainingData))
print(nrow(testData))
```

```r
# g. Thực hiện Linenear Regression với train data. 
```

```r
lmMod <- lm(sales~youtube, data =trainingData)
```

```r
iPred <- predict(lmMod, testData)
```

```r
# i. Dự đoán y_test_predict từ test data => so sánh y_test_pred với y_test 
```

```r
result <- data.frame(testClass = testData$sales[30:40],iPred[30:40])
print(result)
```

```r
# j. Tính Mean Square Error (mse), r^2 cho train, r^2 cho test. Nhận xét. 
```

```r
mse = mean(lmMod$residuals^2)
print(paste("mse: ", mse))
```

```r
mse_test = mean((testData$sales - iPred)^2)
print(paste('mse in test: ', mse_test))
```

```r
summary(lmMod)$r.square
```

```r
# R^2 có giá trị thấp, model fits ~ 62% data => model không tốt
```

```r
# h. In summary của model 
```

```r
print(summary(lmMod))
```

```r
# bk. Tìm Coefficients, Intercept 
```

```r
print(coef(lmMod))
```

```r
beta_youtube  = coef(lmMod)['youtube']
print(paste("slope: ", beta_youtube))
Intercept <- coef(lmMod)["(Intercept)"]
print(paste("Intercept: ", Intercept))
```

```r
# l. Cho youtube lần lượt: x <- c(100, 200, 300) => dự đoán sales. 
```

```r
x <- c(100, 200, 300)
y <- Intercept + beta_youtube * x
print("Solution 1 - result: ")
print(y)
```

```r
# Visualization
plot(input$youtube, input$sale, 
     main = "Sales Linear regression",
     xlab = "Youtube", ylab = "Sales",
     pch = 19, frame = FALSE, col= 'blue')
lines(x, y, col= 'red', type='p')
abline(lmMod, input, col = "green")
legend("topleft", 
       c("dataset", "data_predict"), 
       cex=0.8, 
       fill = c("blue", "red"))
```

# Yêu cầu 2

```r
input1 <- data[,c('sales','youtube','facebook')]
head(input1)
```

```r
par(mfrow = c(1,2))
boxplot(input1$sales, main = 'sales', sub = paste('Outlier rows:', boxplot.stats(input1$sales)$out))
boxplot(input1$youtube, main = 'youtube', sub = paste('Outlier rows:', boxplot.stats(input1$youtube)$out))
boxplot(input1$facebook, main = 'facebook', sub = paste('Outlier rows:', boxplot.stats(input1$facebook)$out))
```

```r
# Không có outlier
```

```r
print(cor(input1$sales, input1$youtube))
print(cor(input1$sales, input1$facebook))
```

```r
set.seed(42)
trainingRowIndex <- sample(1:nrow(input1),  0.7*nrow(input1))
print("Selected training row index:")
print(trainingRowIndex)
trainingData <- input1[trainingRowIndex,]
testData <- input1[-trainingRowIndex,]
print("Rows of training data and test data:")
print(nrow(trainingData))
print(nrow(testData))
```

```r
lmMod <- lm(sales~youtube+facebook, data =trainingData)
```

```r
mse = mean(lmMod$residuals^2)
print(paste("mse: ", mse))
```

```r
mse_test = mean((testData$sales - iPred)^2)
print(paste('mse in test: ', mse_test))
```
