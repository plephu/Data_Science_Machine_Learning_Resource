# Histogram

```r
p = ggplot(iris, aes(Petal.Length))
```

```r
p1 = p + 
geom_histogram(bins = 10, color = 'white', fill = 'blue') +
ggtitle('Distribution of Petal.Length') +
theme_bw(base_size = 12)
```

```r
p2 = p + 
geom_histogram( aes(y=..density..),bins = 10, color = 'white', fill = 'blue') +
geom_density(col= 'red') +
ggtitle('Distribution of Petal.Leghth') +
theme_bw(base_size = 12)
```

```r
grid.arrange(p1,p2,ncol=2)
```

```r
p1 = ggplot(iris, aes(Petal.Length,fill = Species)) +
geom_histogram(position = 'dodge', bins = 30) +
theme_bw(base_size = 14)
```

```r
p1 = ggplot(iris, aes(Petal.Length,fill = Species, color = Species)) +
geom_histogram(position = 'dodge', bins = 30) +
geom_density(alpha = 01) +
theme_bw(base_size = 14)
```

```r
grid.arrange(p1,p2,nrow=2)
```

### Scatter plot

```r
ip = ggplot(data = iris, aes(x= Petal.Width, y = Petal.Length))
```

```r
ip1 <- ip + geom_point() +
theme_bw(base_size =  14)
```

```r
ip2 <- ip + geom_point() +
geom_smooth(method = 'loess') +
theme_bw(base_size =  14)
```

```r
library('gridExtra')
```

```r
grid.arrange(ip1, ip2, ncol =2)
```

```r
ip = ggplot(data = iris, aes(x= Petal.Width, y = Petal.Length, col=Species,fill = Species))
```

```r
ip1 <- ip + geom_point() +
theme_bw(base_size =  14)
```

```r
ip2 <- ip + geom_point() +
geom_smooth(method = 'lm') +
theme_bw(base_size =  14)
```

```r
grid.arrange(ip1, ip2, nrow =2)
```
